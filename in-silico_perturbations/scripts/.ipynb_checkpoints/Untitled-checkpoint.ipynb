```python
from ast import AsyncFunctionDef
import pandas as pd
import numpy as np 
import OneCC_bool
import networkx as nx

import seaborn as sns
import pickle
import matplotlib.pyplot as plt

import os
import scanpy as sc
import multiprocessing as mp
```

```python
data_types = os.listdir("../output/curated_networks")
data_types = [x for x in data_types if 'DS_Store' not in x]
for data_type in data_types:
    state_dict = pickle.load(open("../Beeline_benchmark/run_OneCC_pyEpoch/" + data_type + "/state_dict.pickle", "rb"))
    init_state = state_dict['lineage_0'].iloc[:, 0]

    exp_dict = dict()
    for gene in init_state.index: 
        if init_state[gene] == 1:
            exp_dict[gene] = 2
        else:
            exp_dict[gene] = 0

    GRN_methods = os.listdir("../output/curated_networks/" + data_type)
    GRN_methods = [x for x in GRN_methods if 'DS_Store' not in x]
    for GRN_method in GRN_methods:
        print(GRN_method)
        for network_type in ['orphan', 'density']:
            save_folder_path = "../output/simulation_OneCC/" + data_type + "/" + GRN_method + "/" + network_type
            # insert the bottom code here.
            sim_files = [x for x in os.listdir(save_folder_path + "/simulation/") if '.csv' in x]
            big_sim_df = pd.DataFrame()
            for sim_file in sim_files: 
                experiment_title = sim_file.replace("_exp.csv", "")
                temp_sim = pd.read_csv(save_folder_path + "/simulation/" + sim_file, index_col = 0)
                temp_sim = temp_sim[temp_sim.columns[::50]]
                temp_sim.columns = experiment_title + "_" + temp_sim.columns
                big_sim_df = pd.concat([big_sim_df, temp_sim], axis = 1)
            train_obj = OneCC_bool.UMAP_embedding_train(big_sim_df)
            UMAP_coord = OneCC_bool.UMAP_embedding_apply(train_obj, big_sim_df)
            UMAP_coord['sim_time'] = [int(x.split("_")[2]) for x in list(UMAP_coord.index)]
            UMAP_coord = pd.concat([UMAP_coord, big_sim_df.T], axis = 1)

            train_obj = OneCC_bool.UMAP_embedding_train(big_sim_df)
            UMAP_coord = OneCC_bool.UMAP_embedding_apply(train_obj, big_sim_df)
            UMAP_coord['sim_time'] = [int(x.split("_")[2]) for x in list(UMAP_coord.index)]
            UMAP_coord = pd.concat([UMAP_coord, big_sim_df.T], axis = 1)

            sns.scatterplot(data=UMAP_coord, x= 'UMAP_1', y='UMAP_2', hue='sim_time').figure.savefig(save_folder_path + "/sim_time_UMAP.png", bbox_inches="tight")
            for temp_gene in big_sim_df.index: 
                print(temp_gene)
                sns.scatterplot(data=UMAP_coord, x= 'UMAP_1', y='UMAP_2', hue=temp_gene).figure.savefig(save_folder_path + "/" + temp_gene + "_sim_time_UMAP.png", bbox_inches="tight")
                plt.show()
            big_sim_df.to_csv(save_folder_path + "/big_sim_df.csv")
            UMAP_coord.to_csv(save_folder_path + "/UMAP_coord.csv")
```

```python
train_exp = big_sim_df
```

```python
from sklearn.decomposition import PCA 
```

```python
training_obs = dict()
training_obs['feature_genes'] = np.array(train_exp.index)
train_exp = train_exp.T
train_exp
```

```python
my_PCA = PCA()
my_PCA.fit(train_exp)
```

```python
training_obs['PCA'] = my_PCA
PCA_features = my_PCA.transform(train_exp)
```

```python
if PCA_features.shape[1] < 9: 
    n_PC = PCA_features.shape[1]
else: 
    n_PC = 9

PCA_features = pd.DataFrame(PCA_features, index = train_exp.index)
PCA_features = PCA_features.iloc[:, 0:n_PC]
PCA_features
```

```python
import umap
my_UMAP = umap.UMAP().fit(PCA_features)
training_obs['UMAP'] = my_UMAP
```
