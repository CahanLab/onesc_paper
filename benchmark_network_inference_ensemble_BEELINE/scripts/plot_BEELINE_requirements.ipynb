```python
import scanpy as sc
import pandas as pd
import numpy as np 
import pickle 
import networkx as nx
import onesc
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
```

```python
sc.set_figure_params(dpi_save = 600, fontsize = 18)
```

```python
# look at dyn-BF
output_path = '../output/OneSC_training_plots/dyn-BF'
sc.settings.figdir = output_path # set the figure output path
adata = sc.read_h5ad("../Beeline_benchmark/run_OneSC/dyn-BF/redefined_adata.h5ad")
train_exp = adata.to_df().T
samp_tab = adata.obs
pt_col = 'pseudoTime'
cluster_col = 'leiden'
sc.pl.umap(adata, color = cluster_col, save = cluster_col + ".png")
```

```python
sc.pl.umap(adata, color = pt_col, save = pt_col + ".png")
```

```python
initial_clusters = ['9']
end_clusters = ['6', '5']

clusters_G = onesc.construct_cluster_network(train_exp, samp_tab, initial_clusters = initial_clusters, terminal_clusters = end_clusters, cluster_col = "leiden", pseudo_col = pt_col)

f = plt.figure()
nx.draw(clusters_G, with_labels = True, node_size=400, node_color='#FFFFFF', font_size = 30, width = 2)
f.savefig(output_path + "/graph.png")
```
